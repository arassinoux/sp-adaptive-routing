```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

import plotly
from plotly.graph_objs import Scatter, Layout

plotly.offline.init_notebook_mode(connected=True)
```

```python
y_random = pd.read_csv('platform/data/regret_RANDOM.csv', header=None)
y_egreedy = pd.read_csv('platform/data/regret_EGREEDY.csv', header=None)
y_softmax = pd.read_csv('platform/data/regret_SOFTMAX.csv', header=None)
y_pursuit = pd.read_csv('platform/data/regret_PURSUIT.csv', header=None)
y_ucb1 = pd.read_csv('platform/data/regret_UCB1.csv', header=None)
y_exp3 = pd.read_csv('platform/data/regret_EXP3.csv', header=None)

x = np.arange(0, 2000)

for index, row in y_random.copy().iterrows():
    for i in x:
        y_random.loc[index, i] = row.loc[:i].mean()
        
for index, row in y_egreedy.copy().iterrows():
    for i in x:
        y_egreedy.loc[index, i] = row.loc[:i].mean()
        
for index, row in y_softmax.copy().iterrows():
    for i in x:
        y_softmax.loc[index, i] = row.loc[:i].mean()

for index, row in y_pursuit.copy().iterrows():
    for i in x:
        y_pursuit.loc[index, i] = row.loc[:i].mean()
        
for index, row in y_ucb1.copy().iterrows():
    for i in x:
        y_ucb1.loc[index, i] = row.loc[:i].mean()

for index, row in y_exp3.copy().iterrows():
    for i in x:
        y_exp3.loc[index, i] = row.loc[:i].mean()


data = [
    Scatter(
        x=x,
        y=y_random.mean(),
        name='random',
        error_y=dict(
        array=y_random.std(),
        thickness=1,
        width=0,
        opacity=0.1
        )
    ),
    Scatter(
        x=x, 
        y=y_egreedy.mean(),
        name='e-greedy',
        error_y=dict(
        array=y_egreedy.std(),
        thickness=1,
        width=0,
        opacity=0.1
        )
    ),
    Scatter(
        x=x,
        y=y_softmax.mean(),
        name='softmax',
        error_y=dict(
        array=y_softmax.std(),
        thickness=1,
        width=0,
        opacity=0.1
        )
    ),
    Scatter(
        x=x,
        y=y_pursuit.mean(),
        name='pursuit',
        error_y=dict(
        array=y_pursuit.std(),
        thickness=1,
        width=0,
        opacity=0.1
        )
    ),
    Scatter(
        x=x,
        y=y_ucb1.mean(),
        name='ucb1',
        error_y=dict(
        array=y_ucb1.std(),
        thickness=1,
        width=0,
        opacity=0.1
        )
    ),
    Scatter(
        x=x,
        y=y_exp3.mean(),
        name='exp3',
        error_y=dict(
        array=y_exp3.std(),
        thickness=1,
        width=0,
        opacity=0.1
        )
    )
]
# IPython notebook
plotly.offline.iplot(data, filename='pandas/basic-line-plot')
```
