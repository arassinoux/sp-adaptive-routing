# Notebook for plotting regret per turn

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

import plotly
from plotly.graph_objs import Scatter, Layout, Figure

plotly.offline.init_notebook_mode(connected=True)
```

## Regret per turn

```python
x = np.arange(0,501)
data = []
y = pd.read_csv('platform-official/data/regret/regret.csv', header=None)

for index_1 in y[0].unique():
    line = y.copy().loc[y[0] == index_1]
    line.drop(line.columns[[0,-1]], axis=1, inplace=True)
    for index, row in line.copy().iterrows():
        for i in x:
            line.loc[index, i] = row.loc[:i].mean()
            
        
    data.append(
            Scatter(
                x=x, 
                y=line.mean(),
                name=index_1,
                error_y=dict(
                array=line.std(),
                thickness=1,
                width=0,
                opacity=0.1
                )
            ))
    
layout = Layout(
    title='',
    xaxis=dict(
        title='Round',
        
    ),
    yaxis=dict(
        title='Regret',
        
    )
)
```

```python
len(data)
```

```python
# IPython notebook
fig = Figure(data=data, layout=layout)
plotly.offline.iplot(fig, filename='pandas/basic-line-plot')
```

```python
x = np.arange(1, 501)
data = []
y = pd.read_csv('platform-official/data/reward/reward.csv', header=None)

for index_1 in y[0].unique():
    line = y.copy().loc[y[0] == index_1]
    line.drop(line.columns[[0,-1]], axis=1, inplace=True)
    for index, row in line.copy().iterrows():
        for i in x:
            line.loc[index, i] = row.loc[:i].sum()
    data.append(
            Scatter(
                x=x, 
                y=line.mean(),
                name=index_1,
                error_y=dict(
                array=line.std(),
                thickness=1,
                width=0,
                opacity=0.1
                )
            ))

layout = Layout(
    title='',
    xaxis=dict(
        title='Round',
        
    ),
    yaxis=dict(
        title='Reward',
        
    )
)

# IPython notebook
fig = Figure(data=data, layout=layout)
plotly.offline.iplot(fig, filename='pandas/basic-line-plot')
```
